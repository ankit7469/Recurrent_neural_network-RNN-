```python
# Predict next word in a sequence (Simple RNN)

import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

# step 1: Take data in sequential
X=np.array([
    [0,1,2,3],
    [1,2,3,4],
    [2,3,4,5],
    [3,4,5,6],
    [4,5,6,7]
])
y=np.array([4,5,6,7,8])

# Step 2: Reshape data to fit in RNN
X=X.reshape((X.shape[0],X.shape[1],1))
print("Input shape",X.shape)

# step 3: Build simple RNN model
model=keras.Sequential([
    layers.SimpleRNN(32,activation='tanh',input_shape=(4,1)),
    layers.Dense(1)
])

# Step 4: compile and fit 
model.compile(optimizer='adam',loss='mse')
history=model.fit(X,y,epochs=200,verbose=0)

# Step 5: predict the next number
test_input=np.array([5,6,7,8]).reshape((1,4,1))
pred=model.predict(test_input)
print("The next number is",pred[0][0])
                   
```
